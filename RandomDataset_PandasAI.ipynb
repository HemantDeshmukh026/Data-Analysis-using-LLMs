## Using PandasAI on randomly generated dataset.


```python
### Generating dataset 
import numpy as np
import pandas as pd
from random import choices 
```

```python
np.random.seed(10)
```

```python
num_of_records = 1000
num_of_outlets = 10
num_of_products = 20
```

```python
# Two Quarters (Jan - March, April - June)
dates = pd.date_range(start='2023-01-01', end='2023-06-30')
dates = choices(dates, k=num_of_records)
```

```python
outlets = ['Outlet_'+str(i+1) for i in range(num_of_outlets)]
outlets = choices(outlets, k=num_of_records)
```

```python
products = ['Product_'+str(i+1) for i in range(num_of_products)]
products = choices(products, k=num_of_records)
```

```python
len(products)
```

```python
units_sold=np.random .randint(1,200, num_of_records)
price_per_unit=np.random.uniform(10,500, num_of_records )
```

```python
total_sales=units_sold*price_per_unit
total_sales
```

```python
df = pd.DataFrame({
'Data': dates,
'Outlet':outlets,
'Product': products,
'Unit_Sold':units_sold,
'Price_per_unit': price_per_unit,
'Total_sales': total_sales})
```
